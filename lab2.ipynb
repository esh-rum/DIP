```python
# TASK 1 - Read a grayscale image (given below) and convert the image to 16 levels, then to 4 levels and finally to 1. Display all four images.
img = cv.imread("E:/6th Semester/DIP/Lab/lab2.2.png", 0)
cv.imshow('Image', img)
cv.waitKey(0)

rows, cols = img.shape
img16 = np.zeros((rows, cols), dtype=np.uint8)

for i in range(rows):
    for j in range(cols):
        temp = img[i][j]
        if temp <= 15:
            img16[i][j] = 0
        elif temp > 15 and temp <= 31:
            img16[i][j] = 1 * 16
        elif temp > 31 and temp <= 47:
            img16[i][j] = 2 * 16
        elif temp > 47 and temp <= 63:
            img16[i][j] = 3 * 16
        elif temp > 63 and temp <= 79:
            img16[i][j] = 4 * 16
        elif temp > 79 and temp <= 95:
            img16[i][j] = 5 * 16
        elif temp > 95 and temp <= 111:
            img16[i][j] = 6 * 16
        elif temp > 111 and temp <= 127:
            img16[i][j] = 7 * 16
        elif temp > 127 and temp <= 143:
            img16[i][j] = 8 * 16
        elif temp > 143 and temp <= 159:
            img16[i][j] = 9 * 16
        elif temp > 159 and temp <= 175:
            img16[i][j] = 10 * 16
        elif temp > 175 and temp <= 191:
            img16[i][j] = 11 * 16
        elif temp > 191 and temp <= 207:
            img16[i][j] = 12 * 16
        elif temp > 207 and temp <= 223:
            img16[i][j] = 13 * 16
        elif temp > 223 and temp <= 239:
            img16[i][j] = 14 * 16
        elif temp > 239 and temp <= 255:
            img16[i][j] = 15 * 16
        else:
            pass
        
cv.imshow('Image 16 Levels', img16)
cv.waitKey(0)


img4 = np.zeros((rows, cols), dtype=np.uint8)

for i in range(rows):
    for j in range(cols):
        temp = img[i][j]
        if temp <= 63:
            img4[i][j] = 0
        elif temp > 63 and temp <= 127:
            img4[i][j] = 1 * 64
        elif temp > 127 and temp <= 191:
            img4[i][j] = 2 * 64
        elif temp > 191 and temp <= 255:
            img4[i][j] = 3 * 64
        else:
            pass
        
cv.imshow('Image 4 Levels', img4)
cv.waitKey(0)


img1 = np.zeros((rows, cols), dtype=np.uint8)

for i in range(rows):
    for j in range(cols):
        temp = img[i][j]
        if temp <= 127:
            img1[i][j] = 0
        elif temp > 127 and temp <= 255:
            img1[i][j] = 1 * 128
        else:
            pass
        
cv.imshow('Image 1 Levels', img1)
cv.waitKey(0)
        
        
        
```

```python
def find(label_mat, x):
    """Finds the root label of x with path compression."""
    if label_mat[x] != x:
        label_mat[x] = find(label_mat, label_mat[x])  # Path compression
    return label_mat[x]


def cca(image, target_pixel, value):
    rows, cols = image.shape
    new_img = np.zeros((rows, cols), dtype=np.uint8) 
    label_mat = {}
    val = value

    for i in range(rows):
        for j in range(cols):
            temp = image[i][j]
            
            left = new_img[i][j - 1] if j > 0 else 0
            top = new_img[i - 1][j] if i > 0 else 0
            top_left = new_img[i - 1][j - 1] if (i > 0 and j > 0) else 0
            top_right = new_img[i - 1][j + 1] if (i > 0 and j < cols - 1) else 0
            
            neighbors = [left, top, top_left, top_right]
            labeled_neighbors = [n for n in neighbors if n > 0]
            
            if temp == 0:
                new_img[i][j] = 0
            elif temp == target_pixel:
                if not labeled_neighbors:
                    val += 1
                    label_mat[val] = val  
                    new_img[i][j] = val
                else:
                
                    smallest_label = min(labeled_neighbors)
                    new_img[i][j] = smallest_label

                    for neighbor in labeled_neighbors:
                        root_a = find(label_mat, smallest_label)
                        root_b = find(label_mat, neighbor)
                        if root_a != root_b:
                            label_mat[root_b] = root_a

    for i in range(rows):
        for j in range(cols):
            if new_img[i][j] > 0:
                new_img[i][j] = find(label_mat, new_img[i][j]) 
                    
    cv.imshow('final', new_img)  
    cv.waitKey(0)
    
    unique_labels = set(label_mat.values()) - {0}  # Remove background (0)
    print(f"Number of objects detected: {len(unique_labels)}")
    print("Unique labels after correction:", unique_labels)
```
